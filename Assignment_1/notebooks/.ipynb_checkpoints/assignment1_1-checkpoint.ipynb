`Import Libraries`

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

`Load the image from the relative path`

```python
image_path = os.path.join("..", "images", "pic1.png")  # path to the image
```

`Convert BGR to RGB`

```python
img_bgr = cv2.imread(image_path)  # Read image in BGR format
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  # Convert to RGB
```

`Display Original Image`

```python
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
```

## Applying Image Processing Filters For Image Cartoonifying

### Generating a black-and-white sketch

`Converting RGB Image to Grayscale`

```python
# Convert the image to grayscale
grayscale_image = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
```

```python
# Display original and grayscale images 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(grayscale_image , cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
```

`Noise Reduction Using Median Filter`

```python
# Apply a median filter with a specified kernel size 5x5 to the grayscale image
median_filtered_image = cv2.medianBlur(grayscale_image, 5)
```

```python
# Display original grayscale images and Smoothed Grayscale Image 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(grayscale_image , cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(median_filtered_image , cmap='gray')
plt.title("Smoothed Grayscale Image")
plt.axis("off")
```

`Edge Detection Using Laplacian Filter`

```python
# edge detection using laplacian filter
laplacian_filtered_image = cv2.Laplacian(median_filtered_image, -1, ksize=5, scale=0.6) # scale 0.6 to enhance the edges(smoothed image)
```

```python
# Display Smoothed Grayscale Image and Edge Detection
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(median_filtered_image , cmap='gray')
plt.title("Smoothed Grayscale Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(laplacian_filtered_image , cmap='gray')
plt.title("Edge Detection")
plt.axis("off")
```

`Edges Thresholding`

```python
# Apply a threshold to the edge image
_, binary_image_threshold = cv2.threshold(laplacian_filtered_image, 100, 255, cv2.THRESH_BINARY_INV)
```

```python
# Display Edge Detection and 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(laplacian_filtered_image , cmap='gray')
plt.title("Output from Laplacian Filter")
plt.axis("off")


plt.subplot(1, 2, 2)
plt.imshow(binary_image_threshold , cmap='gray')
plt.title("Output after image thresholding")
plt.axis("off")
```

### Generating a color painting and a cartoon

`Applying Bilateral Filter`

```python
# Apply bilateral filter to the original image
bilateral_img = cv2.bilateralFilter(img_rgb, d=9, sigmaColor=7, sigmaSpace=2)
# d=9 (Balanced smoothing) - diameter of the pixel neighborhood 
# sigmaColor (Medium (75-150)) - filter sigma in the color space
# sigmaSpace (Medium (75-150)) - filter sigma in the coordinate space

# Perform multiple iterations of bilateral filtering to enhance the image
for i in range(30):
    
    bilateral_img = cv2.bilateralFilter(bilateral_img, d=9, sigmaColor=7, sigmaSpace=2)
```

```python
# Display original and grayscale images 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(bilateral_img)
plt.title("Output from bilateral filter")
plt.axis("off")
```

`Creating Cartoon Effect`

```python
# Convert edge mask to 3-channel format to match the image
binary_image_threshold_3_channel = cv2.cvtColor(binary_image_threshold, cv2.COLOR_GRAY2RGB)
```

```python
# Overlay the edge mask on the filtered image
cartoonized_image = cv2.bitwise_and(bilateral_img, binary_image_threshold_3_channel)
```

```python
# Display original and grayscale images 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cartoonized_image)
plt.title(" Final Output")
plt.axis("off")
```
